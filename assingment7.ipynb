### Move intensity of a grayscale image left, right and into a specific range and check its effect on histogram. An example output is attached.


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
bgr_img = cv2.imread('village.jpeg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
grayscale = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale, cmap='gray')
plt.show()
```

```python
grayscale.max() - 50
```

```python
left_shift_intensity_img = grayscale - np.ones(grayscale.shape, dtype=np.uint8)*100
left_shift_intensity_img[left_shift_intensity_img < 0] = 0
left_shift_intensity_hist = cv2.calcHist([left_shift_intensity_img], channels=[0], mask=None, histSize=[255], ranges=[0,255])
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].imshow(left_shift_intensity_img, cmap='gray')
ax[0].set_title('Left shifted Img')

ax[1].plot(left_shift_intensity_hist)
ax[1].set_title('Left shifted intensity histogram')

plt.suptitle('Left shifted intensity and its histogram', fontsize=20)
plt.show()
```
