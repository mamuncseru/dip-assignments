```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
bgr_img = cv2.imread('village.jpeg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
gray_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
pixel_list = []
for i in range(gray_img.shape[0]):
    for j in range(gray_img.shape[1]):
        pixel_list.append(np.binary_repr(gray_img[i, j], width=8))

```

```python
eight_bit = np.reshape(np.array([int(pixel[0])for pixel in pixel_list], dtype=np.uint8)*2**7, newshape=gray_img.shape)
seven_bit = np.reshape(np.array([int(pixel[1])for pixel in pixel_list], dtype=np.uint8)*2**6, newshape=gray_img.shape)
six_bit = np.reshape(np.array([int(pixel[2])for pixel in pixel_list], dtype=np.uint8)*2**5, newshape=gray_img.shape)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14, 6))
ax[0].imshow(eight_bit, cmap='gray')
ax[1].imshow(seven_bit, cmap='gray')
ax[2].imshow(six_bit, cmap='gray')
plt.show()
```
